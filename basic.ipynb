```python
matches.drop_duplicates(inplace=True)
deliveries.drop_duplicates(inplace=True)
```

```python
print(matches.columns)
```

```python
matches['winner'].fillna('No Result', inplace=True)
matches['player_of_match'].fillna('No Award', inplace=True)
```

```python
# Check if nulls are filled
print(matches[['winner', 'player_of_match']].isnull().sum())
```

```python
deliveries.dropna(inplace=True)
```

```python
# Convert date column to datetime format
matches['date'] = pd.to_datetime(matches['date'], dayfirst=True)
```

```python
print(matches.dtypes['date'])
```

```python
# Preview after conversion
print(matches[['date']].head())
```

```python
print("\nCleaned Matches Dataset Info:")
print(matches.info())

print("\nCleaned Deliveries Dataset Info:")
print(deliveries.info())
```

```python
# Matches Dataset
print(matches.shape)
print(matches.columns)
print(matches['Season'].value_counts())
print(matches['winner'].value_counts())
```

```python
# Deliveries Dataset
print(deliveries.shape)
print(deliveries.columns)
print(deliveries['batsman'].value_counts().head(10))
```

```python
# Merge datasets
df = deliveries.merge(matches, left_on='match_id', right_on='id')
```

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Filter out deliveries where a wicket fell
wickets = deliveries[deliveries['dismissal_kind'].notna()]
top_bowlers = wickets['bowler'].value_counts().head(10)
```

```python
team_wins = matches['winner'].value_counts()
total_matches = matches['team1'].value_counts() + matches['team2'].value_counts()
win_percentage = (team_wins / total_matches * 100).sort_values(ascending=False)
```

```python
toss_winner = matches['toss_winner']
match_winner = matches['winner']
toss_decision = matches['toss_decision']

# How often toss winner also won the match
toss_match_win = (toss_winner == match_winner).sum() / len(matches) * 100
```

```python
venue_wins = matches['venue'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

# Example: Top run scorers
top_batsmen.plot(kind='bar', figsize=(10,6), title='Top 10 Run Scorers')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```
